Una vez visto el ***Módulo 1: Fundamentos de Programación***, se presenta la siguiente actividad para que el alumnado pueda poner a prueba lo aprendido durante este módulo.

Dentro de la siguiente dinámica, realizar un **menú CRUD** donde el tema a emplear es libre. Cada función deberá ejecutar una **operación del CRUD**:

*   **Ver todos los registros**
*   **Crear nuevo registro**
*   **Consultar registro**
*   **Actualizar registro**
*   **Eliminar registro**
*   **Salir menú**

```python
# Menú CRUD para inventario de laboratorio
# base de datos

inventario = [
    {"ID":1, "Reactivo": "Glucosa", "Presentación": "500g", "Marca": "Sigma"},
    {"ID":2, "Reactivo": "Etanol", "Presentación": "1000 mL", "Marca": "Sigma"},
    {"ID":3, "Reactivo": "Bicarbonato", "Presentación": "500g", "Marca": "Baker"}
]




# definir función para ver los registros del inventario
def ver_todos_los_registros():
  if len(inventario) == 0:
    print("No hay registros en el inventario")
    return
  for i, item in enumerate(inventario, start=1):
    print(f"\nRegistro {i}:")
    for k, v in item.items():
            print(f"  {k}: {v}")



# definir función para crear registro
def crear_registro():
  ID = int(input("Ingresa el ID: "))
  Reactivo = input("Ingresa el nombre del reactivo: ")
  Presentación = input("Ingresa la presentación del reactivo: ")
  Marca = input("Ingresa la marca del reactivo: ")
  registro = {"ID": ID, "Reactivo": Reactivo, "Presentación": Presentación, "Marca": Marca}
  inventario.append(registro)
  print("Registro creado con éxito")



# definir función para consultar registro
def consultar_registro():
  nombre = input("Ingrese el nombre del reactivo a consultar: ")
  for i, item in enumerate(inventario):
    if item["Reactivo"] == nombre:
      print("Reactivo encontrado")
      for k, v in item.items():
            print(f"  {k}: {v}")
      break
  else:
    print("Reactivo no encontrado")




# definir función para actualizar registro

def menu_actualizar():
    print("Menú Actualizar")
    print("1. Actualizar ID")
    print("2. Actualizar Reactivo")
    print("3. Actualizar Presentación")
    print("4. Actualizar Marca")
    opcion = input("Ingrese la opción a actualizar: ")
    return opcion


def actualizar_ID(i):
    nuevo_ID = int(input("Ingrese el nuevo ID: "))
    inventario[i]["ID"] = nuevo_ID
    print("ID actualizado")

def actualizar_Reactivo(i):
    nuevo_Reactivo = input("Ingrese el nuevo nombre del reactivo: ")
    inventario[i]["Reactivo"] = nuevo_Reactivo
    print("Reactivo actualizado")

def actualizar_Presentacion(i):
    nueva_Presentación = input("Ingrese la nueva presentación del reactivo: ")
    inventario[i]["Presentación"] = nueva_Presentación
    print("Presentación actualizada")

def actualizar_Marca(i):
    nueva_Marca = input("Ingrese la nueva marca del reactivo: ")
    inventario[i]["Marca"] = nueva_Marca
    print("Marca actualizada")

def actualizar_registro():
  nombre = input("Ingrese el nombre del reactivo a actualizar: ")
  for i, item in enumerate(inventario):
    if item ["Reactivo"] == nombre:
      print("Reactivo encontrado")
      for k, v in item.items():
          print(f"  {k}: {v}")
      break
  else:
    print("Reactivo no encontrado")
    return

  while True:
    opcion = menu_actualizar()

    if opcion == "1":
      actualizar_ID(i)
    elif opcion == "2":
      actualizar_Reactivo(i)
    elif opcion == "3":
      actualizar_Presentacion(i)
    elif opcion == "4":
      actualizar_Marca(i)
    elif opcion == "5":
      print("Saliendo del menú")
      break
    else:
      print("Opción no encontrada")



# definir función para eliminar registro
def eliminar_registro():
  nombre = input("Ingrese el nombre del reactivo a eliminar: ")
  for i, item in enumerate(inventario):
      if item["Reactivo"] == nombre:
         del inventario[i]
         print("Reactivo eliminado")
         break
  else:
    print("Reactivo no encontrado")


# Al final ponemos el menú CRUD
# definir la función para mostrar el menú
def menu_CRUD():
  while True:
    print("Menú CRUD")
    print("1. Ver todos los registros")
    print("2. Crear nuevo registro")
    print("3. Consultar registro")
    print("4. Actualizar registro")
    print("5. Eliminar registro")
    print("6. Salir")

    opcion = input("Ingrese la opción deseada: ")

    if opcion == "1":
      ver_todos_los_registros()
    elif opcion == "2":
      crear_registro()
    elif opcion == "3":
      consultar_registro()
    elif opcion == "4":
      actualizar_registro()
    elif opcion == "5":
      eliminar_registro()
    elif opcion == "6":
      print("Saliendo del menú")
      break
    else:
      print("Opción no encontrada")

menu_CRUD()
```
